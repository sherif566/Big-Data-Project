## Sentiment Analysis for the movie reviews

```python
df5['reviews'][0]
```

```python
#remove duplicate reviews in every list of reviews about a movie
df5['reviews'] = df5['reviews'].apply(lambda x: list(set(x)))
```

```python
df5['reviews'][0]
```

```python
df5['reviews'] = df5['reviews'].apply(lambda x: ' '.join(x))
```

```python
df5['reviews'][0]
```

```python
pip install nltk
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

reviews = df5['reviews'] 
sentiment_scores = [sid.polarity_scores(review)['compound'] for review in reviews]
df5['sentiment_score'] = sentiment_scores
```

```python
df5.head(10)
```

```python
positive_threshold = 0.2
negative_threshold = -0.2
sentiment_categories = []
for score in sentiment_scores:
    if score > positive_threshold:
        sentiment_categories.append('Positive')
    elif score < negative_threshold:
        sentiment_categories.append('Negative')
    else:
        sentiment_categories.append('Neutral')
df5['sentiment_category'] = sentiment_categories
```

```python
df5.head(10)
```

```python
from matplotlib import pyplot as plt 
import matplotlib
df5['sentiment_category'].value_counts().plot(kind='bar')
plt.title("Sentiment Analysis Results")
```

## Linear Regression to Predict Movies' Revenues

```python
df5_copy = df5.copy()
df5_copy.genre = df5_copy.genre.str.replace(',', '')
df5_copy.genre = df5_copy.genre.str.split(' ')
```

```python
unique_genres = df5_copy['genre'].explode().unique()
unique_genres
```

```python
for genre in unique_genres:
    temp = []
    for row in df5_copy['genre']:
        if genre in row:
            temp.append(1)
        else:
            temp.append(0)
    df5_copy[genre] = temp
```

```python
pd.set_option('Display.max_columns', None)
df5_copy.head()
```

```python
categorical_features = df5_copy.select_dtypes(include= "object").columns
categorical_features
```

```python
from sklearn.preprocessing import LabelEncoder
#looping through all categories and trnsforming them into numeric values
le = LabelEncoder()
for feature in categorical_features:
  df5_copy[feature] = le.fit_transform(df5_copy[feature].astype(str))
```

```python
corr = df5_copy.corr()
a = corr['total_revenue'].sort_values(ascending=False)[:]
a
```

```python
X = df5_copy[['vote_count', 'runtime', 'vote_average', 'production_countries', 'production_companies','release_date', 
              'genre', 'original_language', 'popularity', 'director','War', 'Mystery', 'thriller', '', 'Drama', 'Adventure',
              'Comedy','Romance', 'Action', 'Documentary', 'Kids', 'family', 'Horror',
              'Sci-fi', 'Biography', 'Fantasy', 'Animation', 'Western', 'Crime',
              'History', 'Musical', 'Lgbtq+', 'Holiday', 'Music', 'Anime',
              'Foreign', 'Gay', 'lesbian']]
Y = df5_copy['total_revenue']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
Y_pred = lr.predict(X_test)
#R-squared of 0.52 means that the predictors explain about 52% of the variation in our response variable.
r2_score (Y_test, Y_pred)
```

## Correlation Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract the relevant columns for movie population and movie_release_year
movie_popularity = df8['movie_popularity']
movie_release_year = df8['movie_release_year']

# Calculate the correlation coefficient
correlation = movie_popularity.corr(movie_release_year)

# Plotting the data as a scatter plot
plt.scatter(movie_release_year, movie_popularity)

plt.xlabel('movie_release_year')
plt.ylabel('movie_popularity')
plt.title('Correlation Between movie_popularity  and movie_release_year (Correlation: {:.2f})'.format(correlation))

plt.show()
```

## Which director has the highest average movie popularity?

```python
# Group the dataframe by director and calculate the mean popularity score for each group
director_popularity = df8.groupby('director_name')['movie_popularity'].mean()

# Find the director with the highest average popularity
highest_avg_popularity_director = director_popularity.idxmax()

print("The director with the highest average movie popularity is:", highest_avg_popularity_director)
```

## The year with the highest number of english movies released

```python
movie_title_language = 'en'

# Filter the dataset by the specified language
df_filtered = df8[df8['movie_title_language'] == movie_title_language]

# Group the filtered dataframe by year and count the number of movies released in each year
year_counts = df_filtered.groupby('movie_release_year')['movie_title'].count()

# Find the year with the highest number of movies released
highest_year = year_counts.idxmax()

print(f"The year with the highest number of {movie_title_language} movies released is {highest_year} with {year_counts[highest_year]} movies.")
```
